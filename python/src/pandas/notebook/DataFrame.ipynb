try to change DataFram to list of dicts

```python
import string
import random

import pandas as pd
```

```python
# gernerate dataframe
df = pd.DataFrame([{
    'name': ''.join(random.sample(list(string.ascii_lowercase), 5)),
    'email': ''.join(random.sample(list(string.ascii_lowercase), 10)),
    'age': random.randint(18, 30)
} for _ in range(50)])

df
```

```python
# convert to dict
# realize method 1
result = []
for k, v in df.set_index(['name'], drop=True).T.to_dict().items():
    result.append({
        'name': k,
        'data': v
    })
result
```

```python
# realize method 2
def refactory_row(row):
    row['data'] = {
        'email': row['email'],
        'name': row['name']
    }
    return row[['name', 'data']]
    
list(df.apply(refactory_row, axis=1).T.to_dict().values())
```
