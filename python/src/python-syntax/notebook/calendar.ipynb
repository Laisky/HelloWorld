```python
import calendar

from IPython import display
```

```python
c = calendar.TextCalendar(calendar.SUNDAY)
c.prmonth(2015, 4)
```

```python
html = c.formatyear(2015, 4)
display.display_html(display.HTML(html))
```

```python
html = c.formatmonth(2015, 4)

display.display_html(display.HTML(html))
```
